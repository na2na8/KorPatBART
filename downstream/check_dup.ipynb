```python
set_dict = dict(appl_number_dict)
```

```python
def filter_DS_store(folders) :
    if folders[0][0] == '.' :
        folders = folders[1:]
    return folders

def get_undedup(path) :
    dataframe = {
        'application_number' : [],
        'claims' : [],
        'abstract' : [],
        'invention_title' : [],
        'ipc_all' : []
    }
    dedup = []
    first_folders = filter_DS_store(sorted(os.listdir(path))) # A_농업
    
    for first in first_folders :
        first_path = os.path.join(path, first) # A_농업
        second_folders = filter_DS_store(sorted(os.listdir(first_path))) # A_농업/01_농업
        for second in second_folders :
            second_path = os.path.join(first_path, second) # A_농업/01_농업
            third_folders = filter_DS_store(sorted(os.listdir(second_path))) # A_농업/01_농업/011_작물재배업
            for third in third_folders :
                third_path = os.path.join(second_path, third) # A_농업/01_농업/011_작물재배업
                jsons = os.listdir(third_path)
                for j in jsons :
                    json_path = os.path.join(third_path, j)
                    with open(json_path, 'r') as f :
                        new = json.load(f)
                    for item in new['dataset'] :
                        try :
                            if set_dict[item['application_number']] :
#                                 print(item['application_number'])
#                                 print(appl_number_dict[item['application_number']])
                                dedup.append(item['application_number'])
                                continue
                        except KeyError :
                            dataframe['application_number'].append(item['application_number'])
                            dataframe['ipc_all'].append(item['ipc_all'])
                            dataframe['invention_title'].append(item['invention_title'])
                            if 'claims' in item :
                                dataframe['claims'].append(item['claims'])
                            else :
                                dataframe['claims'].append(None)
                            if 'abstract' in item :
                                dataframe['abstract'].append(item['abstract'])
                            else :
                                dataframe['abstract'].append(None)
    print(len(dedup))
    return pd.DataFrame(dataframe)
        
```

```python
t = get_undedup(train_folder)
v = get_undedup(valid_folder)
```

```python
def get_undedup(dataframe) :
    new_dataframe = {
        'application_number' : [],
        'invention_title_eng' : [],
        'invention_title_kor' : [],
        'astrt_cont_eng' : [],
        'astrt_cont_kor' : [],
        'claim_eng' : [],
        'claim_kor' : []
    }
    dedup = []
    for idx in range(len(dataframe)) :
        try :
            if set_dict[dataframe['application_number'].iloc[idx]] :
                dedup.append(dataframe['application_number'].iloc[idx])
                continue
        except KeyError :
            new_dataframe['application_number'].append(dataframe['application_number'].iloc[idx])
            new_dataframe['invention_title_eng'].append(dataframe['invention_title_eng'].iloc[idx])
            new_dataframe['invention_title_kor'].append(dataframe['invention_title_kor'].iloc[idx])
            new_dataframe['astrt_cont_eng'].append(dataframe['astrt_cont_eng'].iloc[idx])
            new_dataframe['astrt_cont_kor'].append(dataframe['astrt_cont_kor'].iloc[idx])
            new_dataframe['claim_eng'].append(dataframe['claim_eng'].iloc[idx])
            new_dataframe['claim_kor'].append(dataframe['claim_kor'].iloc[idx])
    print(len(dedup))
    return pd.DataFrame(new_dataframe)
```
